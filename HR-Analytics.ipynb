Performing VIF to treat multicoliniarity found in the data

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_col= df_for_model.select_dtypes(include=['number'])
vif_data = pd.DataFrame()
vif_data['Feature'] = vif_col.columns
vif_data['VIF'] = [variance_inflation_factor(vif_col.values, i) for i in range(vif_col.shape[1])]
vif_data.sort_values(by = 'VIF',ascending=False)
```

Let's Start to drop one at a time starting with the highest one untill no feature exceeds 10 VIF

```python
df_model_vif_treated = df_for_model.copy()
```

```python
vif_col2= df_model_vif_treated.select_dtypes(include=['number'])
```

```python
max_vif = 10

remove_flag = True

while remove_flag:
    vif = pd.DataFrame()
    vif["Feature"] = vif_col2.columns
    vif["VIF"] = [variance_inflation_factor(vif_col2.values, i) for i in range(vif_col2.shape[1])]

    max_vif_feature = vif.loc[vif['VIF'].idxmax()]

    if max_vif_feature['VIF'] > max_vif:
        vif_col2 = vif_col2.drop(max_vif_feature['Feature'], axis=1)
        print(f"Removed variable with high VIF: {max_vif_feature['Feature']} (VIF={max_vif_feature['VIF']})")
    else:
        # If no variable exceeds the threshold, set the flag to False to exit the loop
        remove_flag = False
```

```python
cat_df = df_model_vif_treated.select_dtypes(include=['object'])
df_after_vif = pd.concat([vif_col2, cat_df], axis=1)

df_after_vif
```

                                                Data Spliting and preprocessing

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , OneHotEncoder , StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
#spliting the data
x = df_after_vif.drop('Attrition',axis=1)
y = df_after_vif['Attrition']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
#creating labelencoder
le = LabelEncoder()
col_to_label_encode = ['OverTime','Gender']
for col in col_to_label_encode:
    x_train[col] = le.fit_transform(x_train[col])
    x_test[col] = le.transform(x_test[col])
```

```python
categorical_cols = ["BusinessTravel", "Department", "EducationField", "JobRole", "MaritalStatus"]


ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)

# Fit on train categorical features
x_train_ohe = ohe.fit_transform(x_train[categorical_cols])
x_test_ohe = ohe.transform(x_test[categorical_cols])

# Get the new column names
ohe_cols = ohe.get_feature_names_out(categorical_cols)

# Convert back to DataFrame
x_train_ohe = pd.DataFrame(x_train_ohe, columns=ohe_cols, index=x_train.index)
x_test_ohe = pd.DataFrame(x_test_ohe, columns=ohe_cols, index=x_test.index)

# Concatenate with the rest of the features
x_train_final = pd.concat([x_train.drop(categorical_cols, axis=1), x_train_ohe], axis=1)
x_test_final = pd.concat([x_test.drop(categorical_cols, axis=1), x_test_ohe], axis=1)
```

```python
col_to_scal = ['DailyRate','DistanceFromHome','Education','HourlyRate','MonthlyRate','YearsSinceLastPromotion_transformed']
scale = StandardScaler()
for col in col_to_scal:
    x_train_final[[col]] = scale.fit_transform(x_train_final[[col]])
    x_test_final[[col]] = scale.transform(x_test_final[[col]])
```

                                                    Model Training and Selection

```python
log_reg = LogisticRegression(max_iter=1000, random_state=42)  
log_reg.fit(x_train_final, y_train)

y_train_pred = log_reg.predict(x_train_final)
y_test_pred  = log_reg.predict(x_test_final)

train_acc = accuracy_score(y_train, y_train_pred)
test_acc  = accuracy_score(y_test, y_test_pred)

print(f"Train Accuracy: {train_acc:.4f}")
print(f"Test Accuracy:  {test_acc:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder , OneHotEncoder , StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    auc
)

# NOTE: Placeholder for loading your processed DataFrame
# You must ensure 'df_after_vif' is loaded and defined here before running.
# Example:
# try:
#     df_after_vif = pd.read_csv('your_vif_processed_data.csv')
# except FileNotFoundError:
#     print("ERROR: Please load the 'df_after_vif' DataFrame before executing.")
#     exit()
# Assuming df_after_vif is loaded:

# --- 1. Define Plotting Functions ---

def plot_roc_curve(y_true, y_scores, model_name='Model'):
    """Plots the ROC curve and displays the AUC score."""
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'{model_name} ROC (AUC = {roc_auc:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Guess')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (FPR)')
    plt.ylabel('True Positive Rate (TPR) - Recall')
    plt.title(f'Receiver Operating Characteristic (ROC) Curve - {model_name}')
    plt.legend(loc="lower right")
    plt.grid(True)
    plt.show()

def plot_confusion_matrix(y_true, y_pred, model_name='Model', classes=['Stay', 'Leave']):
    """Plots the confusion matrix."""
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=classes)
    
    # Increase figure size for better visibility
    fig, ax = plt.subplots(figsize=(8, 8))
    disp.plot(cmap=plt.cm.Blues, values_format='d', ax=ax)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.show()

# --- 2. Data Splitting and Preprocessing (Copied from your code) ---

# Spliting the data
x = df_after_vif.drop('Attrition',axis=1)
y = df_after_vif['Attrition']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)

# Label Encoding
le = LabelEncoder()
col_to_label_encode = ['OverTime','Gender']
for col in col_to_label_encode:
    x_train[col] = le.fit_transform(x_train[col])
    x_test[col] = le.transform(x_test[col])

# One-Hot Encoding
categorical_cols = ["BusinessTravel", "Department", "EducationField", "JobRole", "MaritalStatus"]
ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)

x_train_ohe = ohe.fit_transform(x_train[categorical_cols])
x_test_ohe = ohe.transform(x_test[categorical_cols])

ohe_cols = ohe.get_feature_names_out(categorical_cols)

# Convert back to DataFrame, ensuring index alignment
x_train_ohe = pd.DataFrame(x_train_ohe, columns=ohe_cols, index=x_train.index)
x_test_ohe = pd.DataFrame(x_test_ohe, columns=ohe_cols, index=x_test.index)

# Concatenate with the rest of the features
x_train_final = pd.concat([x_train.drop(categorical_cols, axis=1), x_train_ohe], axis=1)
x_test_final = pd.concat([x_test.drop(categorical_cols, axis=1), x_test_ohe], axis=1)

# Scaling
col_to_scal = ['DailyRate','DistanceFromHome','Education','HourlyRate','MonthlyRate','YearsSinceLastPromotion_transformed']
scale = StandardScaler()
for col in col_to_scal:
    x_train_final[[col]] = scale.fit_transform(x_train_final[[col]])
    x_test_final[[col]] = scale.transform(x_test_final[[col]])


# --- 3. Logistic Regression Training ---

log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(x_train_final, y_train)

y_test_pred_lr = log_reg.predict(x_test_final)
y_test_prob_lr = log_reg.predict_proba(x_test_final)[:, 1] # Probabilities for ROC-AUC

train_acc_lr = accuracy_score(y_train, log_reg.predict(x_train_final))
test_acc_lr = accuracy_score(y_test, y_test_pred_lr)

print("--- Logistic Regression Model Results ---")
print(f"Train Accuracy: {train_acc_lr:.4f}")
print(f"Test Accuracy:  {test_acc_lr:.4f}")

# --- 4. KNN Training (For completeness) ---

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_final, y_train)

train_acc_knn = accuracy_score(y_train, knn.predict(x_train_final))
test_acc_knn = accuracy_score(y_test, knn.predict(x_test_final))

print("\n--- KNN Model Results (Baseline) ---")
print(f"KNN Train Accuracy: {train_acc_knn:.4f}")
print(f"KNN Test Accuracy:  {test_acc_knn:.4f}")


# --- 5. Calculate and Display Key Metrics for Logistic Regression ---

# ROC-AUC
roc_auc = roc_auc_score(y_test, y_test_prob_lr)
# Precision, Recall, F1-Score
precision = precision_score(y_test, y_test_pred_lr)
recall = recall_score(y_test, y_test_pred_lr)
f1 = f1_score(y_test, y_test_pred_lr)

print("\n--- Logistic Regression Key Imbalance Metrics (Test Set) ---")
print(f"ROC-AUC (Discriminatory Power): {roc_auc:.4f}")
print(f"Precision (Accuracy of 'Leaver' predictions): {precision:.4f}")
print(f"Recall (Capture Rate of actual Leavers): {recall:.4f}")
print(f"F1-Score (Balanced Measure): {f1:.4f}")

# --- 6. Generate Visualizations for Logistic Regression ---
plot_roc_curve(y_test, y_test_prob_lr, model_name='Logistic Regression')
plot_confusion_matrix(y_test, y_test_pred_lr, model_name='Logistic Regression')
```
